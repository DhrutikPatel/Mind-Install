## implementing dataset

```python
input_col = ['Symptom_1', 'Symptom_2', 'Symptom_3']
X = df.loc[:, input_col]
Y = df.loc[:, 'Disease']
print(X)
print(Y)
```

```python
Xuni = np.unique(X, return_inverse = True)
Xuni[0].shape
```

```python
index = np.where(Xuni[0] == " skin_rash")[0][0]
```

```python
categories = np.unique(Y, return_inverse = True)
categories[0].size
```

```python
y = categories[1]
print(y.shape)
y
```

```python
x = np.zeros((Y.size,3))
for j in range(3):
    i = 0
    for sym in X["Symptom_"+str(j+1)]:
        x[i][j] = np.where(Xuni[0] == sym)[0][0]
        i += 1
        
        
x
```

# K nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x, y)

model.score(x, y)
```
